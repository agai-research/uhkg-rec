# CAGE baseline

Runs the simplified CAGE baseline standalone and shows a sample top-k recommendation.

```python
%cd UHKG-Rec
import sys, os
sys.path.insert(0, os.getcwd())
```

```python
from Baselines.CAGE.cage import run_cage
model, g = run_cage(theta=0.7, seed=0, layers=2)
symptoms = [n for n, d in g.nodes(data=True) if d['type'] == 'S'][:2]
services = [n for n, d in g.nodes(data=True) if d['type'] == 'H']
rec = model.recommend(symptoms, services, k=5)
print('query symptoms:', symptoms)
for h, s in rec:
    print(f'  {h}: {s:.4f}')
```
