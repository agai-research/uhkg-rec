Summarizes the generated dataset (entity counts, confidence-score distribution, conflict/interaction density) and the statistical significance of the experimental comparisons.

## Dataset statistics

## Confidence-score distribution (extracted facts)

```python
import json, matplotlib.pyplot as plt
with open(os.path.join(cfg.RAW_DIR, 'facts.json')) as f:
    facts = json.load(f)
confidences = [f['confidence'] for f in facts]
plt.hist(confidences, bins=20, color='#0072B2', edgecolor='white')
plt.xlabel('confidence score c')
plt.ylabel('number of facts')
plt.title('Distribution of fact confidence scores')
plt.show()
```

## Requirement / service pattern correlation matrix

```python
with open(os.path.join(cfg.DATA_DIR, 'pattern_matrix.json')) as f:
    matrix_info = json.load(f)
print('N_rp =', matrix_info['n_rp'], ' N_sp =', matrix_info['n_sp'])
import numpy as np
m = np.array(matrix_info['matrix'])
plt.imshow(m, aspect='auto', cmap='viridis')
plt.colorbar(label='PatternMatch(RP_i, SP_j)')
plt.xlabel('service pattern index'); plt.ylabel('requirement pattern index')
plt.title('Correlation matrix M^rh')
plt.show()
```
